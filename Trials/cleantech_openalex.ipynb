```python
import os
import pandas as pd
import requests
import json
```

```python
# Get ID of clean technology concept from OpenAlex

cleantech_url = 'https://api.openalex.org/concepts?search=clean technology'
response_cleantech = requests.get(cleantech_url)
response_cleantech.json()
open_alex_cleantech = response_cleantech.json()
```

```python
# Get all works listed under the "Clean technology" concept in OpenAlex

for page in range(6):
    cleantech_works_url = 'https://api.openalex.org/works?filter=concepts.id:' + str(open_alex_cleantech['results'][0]['ids']['openalex']) + ',has_abstract:true&page=' + str(page) + '&sample=1000&per-page=200&seed=42' # "Clean technology" concept
    cleantech_works = requests.get(cleantech_works_url).json()
    # Save cleantech_works dictionary to json
    with open('/Users/juergenthiesen/Documents/cleantech_works' + str(page) + '.json', 'w') as fp:
        json.dump(cleantech_works, fp)
```
